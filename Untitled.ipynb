```python
# Initialize a list to store the inertia values
inertia_values = []

# Loop over different values of k (from 1 to 10)
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X)  # Fit the KMeans model
    inertia_values.append(kmeans.inertia_)  # Append the inertia to the list

# Plot the Elbow Method curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia_values, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Distortion (Inertia)')
plt.grid(True)
plt.show()
```

```python
# Creating the KMeans model with 5 clusters
model = KMeans(n_clusters=5, init='k-means++', random_state=0)

# Fitting the model and predicting the cluster labels
y_kmeans = model.fit_predict(X)

# 'y_kmeans' contains the cluster labels for each data point
print(y_kmeans)
```

```python

# Add the cluster labels to the original DataFrame
df['Cluster'] = y_kmeans
# Plotting the clusters
plt.figure(figsize=(8, 6))

# Scatter plot: Color points by cluster label
plt.scatter(df['Age'], df['Annual Income (k$)'], c=df['Cluster'], cmap='viridis', s=100, alpha=0.6)

# Plot customization
plt.title('KMeans Clusters - Age vs Annual Income')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.colorbar(label='Cluster')  # Add colorbar to show the clusters
plt.grid(True)
plt.show()
```

```python
# Save the KMeans model and DataFrame to a .pkl file
with open('kmeans_model.pkl', 'wb') as model_file:
    pickle.dump(model, model_file)

with open('df_with_clusters.pkl', 'wb') as df_file:
    pickle.dump(df, df_file)

print("Model and DataFrame saved successfully!")
```
